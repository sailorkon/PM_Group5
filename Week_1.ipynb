## View Data

Operation Settings

```python
for i in range(1,4):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.suptitle('Operation Setting %.f' %i)
    ax1.plot(curr_dataSet["time"],curr_dataSet["opSett_"+str(i)])
    ax2.scatter(curr_dataSet["time"],curr_dataSet["opSett_"+str(i)])
    plt.xlabel("time")
```

```python
for i in range(1,21):
    plt.figure(i)
    plt.plot(curr_dataSet["time"],curr_dataSet["measure_"+str(i)])
    plt.ylabel("sensor measurement %.f" %i)
    plt.xlabel("time")
    plt.ylim(min(curr_dataSet["measure_"+str(i)]), max(curr_dataSet["measure_"+str(i)]))
    
```
